# Data load

## Import CSV file into a dataframe

```python
import pandas as pd
path = '../../data/consolidation-etalab-schema-irve-statique-v-2.2.0-20231031.csv'

df = pd.read_csv(path)

print(df.shape)

df.head()
```

```python
# connect to postgres database
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:postgres@localhost:5432/postgres')

# write to postgres database
# df.to_sql('Stations2', engine, if_exists='replace', index=True, index_label='id')
df.to_sql('Stations', engine, if_exists='append', index=True, index_label='id')
```

```python
# count rows in postgres database
connection = engine.connect()

query = text("SELECT COUNT(*) as count FROM irve")
result = connection.execute(query)

for row in result:
    print("count:", row)
```
